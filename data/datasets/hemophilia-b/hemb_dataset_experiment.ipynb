```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python

dataset = pd.read_csv("HemB_Dataset_SENAI_v5a.csv", sep="\t")
label_encoder = LabelEncoder()
dataset['Domain'] = label_encoder.fit_transform(dataset.loc[:,'Domain'])
dataset['Protein_Change'] = label_encoder.fit_transform(dataset.loc[:,'Protein_Change'])
dataset['aa1'] = label_encoder.fit_transform(dataset.loc[:,'aa1'])
dataset['aa2'] = label_encoder.fit_transform(dataset.loc[:,'aa2'])
print(dataset)
```
